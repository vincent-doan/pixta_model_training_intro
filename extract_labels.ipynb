```python
main_data_path = 'pascal_voc_2007_data/voctrainval_06-nov-2007/VOCdevkit/VOC2007/ImageSets/Main/'
# For train & val set: 'pascal_voc_2007_data/voctrainval_06-nov-2007/VOCdevkit/VOC2007/ImageSets/Main/'
# For test set: 'pascal_voc_2007_data/voctest_06-nov-2007/VOCdevkit/VOC2007/ImageSets/Main/'

path_ending = '_train.txt'
# For train set: '_train.txt'
# For val set: '_val.txt'
# For test set: '_test.txt'

output_path = 'train_labels.csv'
# For train set: 'train_labels.csv'
# For val set: 'val_labels.csv'
# For test set: 'test_labels.csv'
```

```python
import os

train_data_paths = [main_data_path + path for path in sorted(os.listdir(main_data_path)) if path.endswith(path_ending) ]
labels = [path.split('/')[-1][:-len(path_ending)] for path in train_data_paths]

train_data_paths
labels
```

```python
from collections import defaultdict

annotation_per_class = defaultdict(list)
for path in train_data_paths:
    with open(path, 'r') as file:
        for line in file:
            image_id, annotation = line[:6], int(line[6:].strip())
            annotation_per_class[image_id].append(annotation)
```

```python
for key, val in annotation_per_class.items():
    print(key, '-->', val)
```

```python
import pandas as pd

df = pd.DataFrame(annotation_per_class)
df
```

```python
df = df.reset_index(drop=True)
df["index"] = labels
df.set_index('index', inplace=True)
df = df.transpose()
df
```

```python
df.to_csv(output_path)
```
