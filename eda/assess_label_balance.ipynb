From documentation on PASCAL Visual Object CLasses Challenge 2007 (VOC2007) Development Kit:</br>

"Participants are free to omit these images (i.e. images labelled as difficult, or having a 0 in its label) from training or include as either positive or negative examples."

In this project, these images are to be set to negative.

```python
import pandas as pd
```

```python
train_labels = pd.read_csv('../labels/train_labels.csv', index_col=0)
labels = train_labels.columns
labels = list(labels)[1:]
labels
```

```python
res = pd.DataFrame([])
for label in labels:
    res = pd.concat([res, train_labels[label].value_counts()], axis=1)

res.columns = labels
res = res.fillna(value=0)
res
```

```python
val_labels = pd.read_csv('../labels/val_labels.csv', index_col=0)
res = pd.DataFrame([])
for label in labels:
    res = pd.concat([res, val_labels[label].value_counts()], axis=1)

res.columns = labels
res = res.fillna(value=0)
res
```

```python
test_labels = pd.read_csv('../labels/test_labels.csv', index_col=0)
res = pd.DataFrame([])
for label in labels:
    res = pd.concat([res, test_labels[label].value_counts()], axis=1)

res.columns = labels
res = res.fillna(value=0)
res
```

```python
train_labels.replace(-1, 0, inplace=True)
train_labels.to_csv('../labels/processed_train_labels.csv')
val_labels.replace(-1, 0, inplace=True)
val_labels.to_csv('../labels/processed_val_labels.csv')
test_labels.replace(-1, 0, inplace=True)
test_labels.to_csv('../labels/processed_test_labels.csv')
```
